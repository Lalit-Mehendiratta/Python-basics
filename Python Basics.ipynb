#### Visualizing data with pandas

```python
#Create a small DataFrame 
import pandas as pd
df = pd.DataFrame(index=['Atiya', 'Abbas', 'Cornelia','Stephanie', 'Monte'],
                   data={'Apples':[20, 10, 40, 20, 50],'Oranges':[35, 40, 25, 19, 33]})
```

```python
df
```

```python
#color=['.2',".7"]
ax = df.plot.bar(color=['.7','.3'], figsize=(16,4))
#ax.get_figure()
```

```python
#Density plot
ax = df.plot.kde(color=['.2',".7"], figsize=(16,4))
```

```python
#Let's plot a line plot, scatter plot, and a bar plot in a single figure
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16,4))
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16,4))
fig.suptitle('Two Variable Plots')
df.plot.line(ax=ax1, title='Line plot')
df.plot.scatter(x='Apples', y='Oranges',ax=ax2, title='Scatterplot')
df.plot.bar(ax=ax3, title='Bar plot')
```

# seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips =sns.load_dataset('tips')
```

```python
tips
```

## Univariate → “one variable” data visualization

```python
#strip plot

sns.stripplot(y= 'tip', data=tips,jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

```python
#Grouping with stripplot()
sns.stripplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.stripplot(x='day', y='tip', data=tips, size=4,jitter=False)
plt.ylabel('tip ($)')
plt.show()
```

```python
#Swarm plot

sns.swarmplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
    
```

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='sex')
plt.ylabel('tip ($)')
plt.show()
```

```python
#Changing orientation
sns.swarmplot(x='tip', y='day', data=tips, hue='sex',orient='h')
plt.xlabel('tip ($)')
plt.show()
```

### Box and Violin plot

```python
#plt.subplot(1,2,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
```

```python
#plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.show()
```

## Bivariate → “two variables” data visualization

```python
#Joint plot

sns.jointplot(x= 'total_bill', y= 'tip', data=tips)
plt.show()
```

```python
# Density plot
sns.jointplot(x='total_bill', y= 'tip', data=tips,kind='kde')
plt.show()
```

```python
#Pair plot

sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips, hue='sex')
plt.show()
```

### Proportion of target (Survived)

```python
data.groupby('Survived')['PassengerId'].count()
```

This dataset has a decent proportion of target class and it is not skewed to any one.

### Visual Exploration

```python
f, ax = plt.subplots(figsize=(11,5))
sns.boxplot(x='Survived', y="Age",  data=data);
```

Very less chances for age > 60+ to survive (except some outliers)

```python
f, ax = plt.subplots(figsize=(7,3))
sns.barplot(x='Sex', y="Survived",  data=data);
```

```python
f, ax = plt.subplots(figsize=(11,5))
sns.boxplot(x="Sex", y="Age", hue="Survived", data=data);
```

Gives clarity to the above finding that Male has very less chances for age > 60+

Number of female passengers survived is more than male passengers

```python
sns.barplot(x="Pclass", y="Survived", data=data);
```

Passenger Class 1 has high survival rate

```python
sns.barplot(x="Pclass", y="Survived",hue="Sex", data=data);
```

```python
sns.barplot(x="SibSp", y="Survived", data=data);
```

More the siblings less the survival chance

```python
sns.barplot(x="Parch", y="Survived", data=data);
```

Those who are occumpanied by a family member (elder or siblings) had high survival rate

Passengers aged between 18-38 had high survival rate compared to others

```python
survived = data.loc[data['Survived']==1,"Age"].dropna()
sns.distplot(survived)
plt.title("Survived");
```

```python
not_survived = data.loc[data['Survived']==0,"Age"].dropna()
sns.distplot(not_survived)
plt.title("Not Survived");
```

Infants had high survival rate and elderly passengers above 65+ were less likely to survive
